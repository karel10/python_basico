A continuación procedamos a obtener un dataframe reducido. Para eso haremos varias cosas:
+ Eliminaremos las filas que contengan registros vacíos ("NaN").
+ Pero ojo, antes, como los strings vacíos no se reconocen como tal, hemos de "traducirlos" a NaN.
+ Luego hacemos, ya sí, la eliminación de filas.
+ Tomamos la muestra de 200000 registros completos (ya ninguno tendrá NaN)
+ Eliminamos columnas redundantes o a las que no podamos dar utilidad.


Tenemos cinco campos que son strings ("object", en nomenclatura Pandas), son los siguientes:
"UniqueCarrier", "TailNum", "Origin", "Dest", "CancellationCode".
Hacemos la substitución de los "strings" vacíos en cada una de esas columnas por "NaN". 

```python
len(retrasos_full)
```

```python
campos_strings = "UniqueCarrier", "TailNum", "Origin", "Dest", "CancellationCode"

for element in campos_strings:  # Sustituimos strings vacíos por NaN
    retrasos_full[element].replace('', np.nan, inplace=True)
```

```python
retrasos_full.dropna(axis=0, inplace=True)  # Eliminamos filas que contienen NaN.
```

Adicionalmente, podemos eliminar filas duplicadas, si es que las hubiera.

```python
retrasos_full.drop_duplicates(inplace=True)
```

```python
len(retrasos_full)  # Podemos comprobar como se han eliminado 689272 registros, aprox. un 35% de la muestra original.
```

Eliminaremos las siguientes columnas:
    
+ "Unnamed: 0", porque es redundante, no nos aporta información, no se sabe muy bien qué es ese número de orden (algún identificador).
+ "Year", porque todos los vuelos son del 2008, no aporta información diferenciadora.
+ "DepTime" y "CRSDepTime" (hora real y prevista de salida), porque servirían para para obtener "DepDelay", pero como ya tenemos este dato, son redundantes.
+ "ArrTime" y "CRSArrTime" (hora real y prevista de llegada) son redudantes por el mismo motivo, si nos damos cuenta de que ya tenemos el campo "ArrDelay".
+ "FlightNum" y "TailNum" solo son identificadores concretos que interpreto que no nos aportarán información sobre las tendencias generales. En cambio, sí nos quedamos con la compañía, por si vemos un patrón (p.ej. compañía con más retrasos que otras).

Pendiente de si al final usamos todas las demás o no, estas pienso que son las más descartables.

```python
retrasos_reducido = retrasos_full.sample(200000)  # Creamos la muestra reducida que pedía el enunciado.
# Nota: después de consultarlo, es que hice el muestreo DESPUÉS de eliminar las filas que contenían campos vacíos.
```

```python
# Eliminamos las columnas especificando sus nombres y "axis = 1":
retrasos_reducido = retrasos_reducido.drop(['Unnamed: 0', 'Year', 'DepTime', 'CRSDepTime', 'ArrTime', 'CRSArrTime', 'FlightNum', 'TailNum'], axis=1)
```

```python
len(retrasos_reducido)
```

```python
retrasos_reducido
```

Liberamos de la memoria el dataframe original, más grande:

```python
del retrasos_full
gc.collect()
retrasos_full=pd.DataFrame()
# Con esto pandas ya no guarda en su memoria el dataframe completo.
```

Como parte del ejercicio, creemos algunas nuevas columnas calculadas a partir de las iniciales. Podemos calcular, por ejemplo: la velocidad media de un vuelo, o la diferencia entre la duración real de un vuelo y la prevista (que viene a ser el tiempo recuperado o perdido desde la salida hasta la llegada.)
Podemos llamarlas:
v_media, dur_real_vs_prevista

```python
retrasos_reducido["v_media"] = retrasos_reducido["Distance"]/(retrasos_reducido["ActualElapsedTime"]/60)
retrasos_reducido["dur_real_vs_prevista"] = retrasos_reducido["ActualElapsedTime"] - retrasos_reducido["CRSElapsedTime"]
```

Para ver qué RUTAS acumulan más número de retrasos, hemos de sumar el retraso de las rutas. En este caso consideraremos como una ruta aquella que transcurre en la misma dirección y sentido (i.e. Honolulu -> Newark diremos que es una ruta "diferente" que la inversa).

```python
# Agrupamos por Origen y Destino.
# Esto da los retrasos acumulados de cada ruta concreta.
retrasos_reducido.groupby(['Origin', 'Dest'])['ArrDelay'].sum()
```

```python
retrasos_reducido.groupby(['Origin', 'Dest'])['ArrDelay'].sum().sort_values(ascending=False)
```
